```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
data = pd.read_csv("../Data/customer_churn_data.csv")
```

```python
# Display first few rows
display(data.head())
```

1️⃣ DESCRIPTIVE STATISTICS 

```python
print("\nBasic Information about Dataset:")
print(data.info())
```

```python
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Checking for missing values
print("\nMissing Values in Dataset:")
print(data.isnull().sum())
```

2️⃣ CHURN DISTRIBUTION

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="ChurnStatus", data=data, hue="ChurnStatus", palette="coolwarm", legend=False)
plt.title("Churn Distribution")
plt.xlabel("Churn Status")
plt.ylabel("Count")
plt.show()
```

3️⃣ AGE DISTRIBUTION

```python
plt.figure(figsize=(8, 5))
sns.histplot(data["Age"], bins=20, kde=True, color="blue")
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

 4️⃣ MONTHLY CHARGES vs CHURN

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="ChurnStatus", y="MonthlyCharges", hue="ChurnStatus", data=data, palette="coolwarm", legend=False)
plt.title("Monthly Charges vs Churn")
plt.show()
```

5️⃣ DATA USAGE vs CHURN

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="DataUsage", y="MonthlyCharges", hue="ChurnStatus", data=data, palette="coolwarm")
plt.title("Data Usage vs Monthly Charges")
plt.show()
```

6️⃣ CALL DURATION ANALYSIS

```python
plt.figure(figsize=(8, 5))
sns.histplot(data["AverageCallDuration"], bins=30, kde=True, color="green")
plt.title("Distribution of Average Call Duration")
plt.xlabel("Call Duration (minutes)")
plt.ylabel("Count")
plt.show()
```

7️⃣ PLAN TYPE BREAKDOWN

```python
plt.figure(figsize=(6, 6))
data["PlanType"].value_counts().plot.pie(autopct="%1.1f%%", colors=["lightblue", "lightcoral"])
plt.title("Plan Type Distribution")
plt.ylabel("")
plt.show()
```

 8️⃣ CORRELATION ANALYSIS

```python
plt.figure(figsize=(10, 6))

# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=['number'])

sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

9️⃣ RELATIONSHIP BETWEEN CALLS & CHURN

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="ChurnStatus", y="NumberOfCalls", data=data, hue="ChurnStatus", legend=False, palette="coolwarm")
plt.title("Number of Calls vs Churn")
plt.show()
```

🔍 INSIGHTS & CONCLUSIONS

```python
print("\n🔹 **Key Insights from the Analysis:**\n")

print("1️⃣ Customers with high monthly charges tend to churn more.")
print("2️⃣ Customers with low data usage have a higher probability of churning.")
print("3️⃣ Basic plan users show a slightly higher churn rate.")
print("4️⃣ Older customers tend to churn less compared to younger customers.")
print("5️⃣ Customers making fewer calls are more likely to churn.")

print("\n📌 **Recommendations to Reduce Churn:**\n")
print("- Offer discounts or loyalty benefits for customers with high monthly charges.")
print("- Provide more data benefits for low-usage customers to encourage engagement.")
print("- Improve customer experience for Basic plan users to retain them.")
print("- Personalized offers for younger customers to increase retention.")
```
