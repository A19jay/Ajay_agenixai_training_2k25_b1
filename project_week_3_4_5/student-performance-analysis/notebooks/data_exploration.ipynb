```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv(r'C:\Users\ajayr\OneDrive\Desktop\agenix\student_performance_analysis\raw\student_performance_data.csv')
```

```python
# Display basic information and summary
print(data.info())
print(data.describe())
```

```python
# Count of students by gender
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=data, hue='Gender', palette='coolwarm', legend=False)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Count of students by Socioeconomic Status
plt.figure(figsize=(6,4))
sns.countplot(x='SocioeconomicStatus', data=data, hue='SocioeconomicStatus', palette='viridis', legend=False)
plt.title('Socioeconomic Status Distribution')
plt.xlabel('Socioeconomic Status')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Grades
plt.figure(figsize=(8,6))
sns.histplot(data['Grades'], bins=20, kde=True, color='blue')
plt.title('Distribution of Grades')
plt.xlabel('Grades')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Attendance
plt.figure(figsize=(8,6))
sns.histplot(data['Attendance'], bins=20, kde=True, color='green')
plt.title('Distribution of Attendance')
plt.xlabel('Attendance Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for Grades by Gender
plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='Grades', data=data, hue='Gender', palette='pastel', dodge=False)
plt.title('Grades by Gender')
plt.xlabel('Gender')
plt.ylabel('Grades')
plt.show()
```

```python
# Boxplot for Grades by Socioeconomic Status
plt.figure(figsize=(6,4))
sns.boxplot(x='SocioeconomicStatus', y='Grades', data=data, hue='SocioeconomicStatus', palette='muted', dodge=False)
plt.title('Grades by Socioeconomic Status')
plt.xlabel('Socioeconomic Status')
plt.ylabel('Grades')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Academic Performance Status count
plt.figure(figsize=(6,4))
sns.countplot(x='AcademicPerformanceStatus', data=data, hue='AcademicPerformanceStatus', palette='Set2', legend=False)
plt.title('Pass vs Fail Count')
plt.xlabel('Academic Performance Status')
plt.ylabel('Count')
plt.show()
```

```python
# Pairplot for numerical features
sns.pairplot(data, hue='AcademicPerformanceStatus', diag_kind='kde', palette='husl')
plt.show()

print("EDA completed successfully.")
```
